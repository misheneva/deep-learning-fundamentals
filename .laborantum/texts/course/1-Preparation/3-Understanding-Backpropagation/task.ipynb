# Task 1

Write function that calculates value of this graph for any given input vector `x` and set of coefficients $b_1, b_2, c_1, c_2$ packed as vector of weights `w`.
In this function also return all the $z$ values.

````python
import numpy as np


## YOU CAN DEFINE ANY FUNCTIONS YOU NEED ##
def sigmoid(z):
    return 1/(1 + np.exp(-z))


def graph_value(x, w):
    y = x
    ## YOUR CODE HERE ##
    x1 = x[0]
    x2 = x[1]
    b1 = w[0]
    b2 = w[1]
    c1 = w[2]
    c2 = w[3]
    z1 = x1 + b1
    z2 = x2 + b2
    z3 = sigmoid(z1)
    z4 = sigmoid(z2)
    z5 = np.tanh(z2)
    z6 = z5 * c2
    z7 = z1 * z4
    z8 = z7 * c1
    z9 = z3 * z6
    y  = z8 + z9
    z  = [z1, z2, z3, z4, z5, z6, z7, z8, z9]
    return y, z
````

````python
answer['graph_value'] = [graph_value(**input) for input in inputs]
````

# Task 2

In this graph, find all direct derivatives of all the values over $z$ or $x$ or $w$.

For example, if $y = z_8 + z_9$, you need to find only $\partial_{z_9} y$ and $\partial_{z_8} y$ for that operation.

You have to do that for all the intermediate signals.

Write the result in form of dictionary, for example, $\partial_{z_9} y$:

```
['dy_dz9'] = 1
```

````python
def graph_derivatives(x, w):
    res = {}

    ## YOUR CODE HERE ##
    x1 = x[0]
    x2 = x[1]
    b1 = w[0]
    b2 = w[1]
    c1 = w[2]
    c2 = w[3]

    y, z = graph_value(x, w)
    z1, z2, z3, z4, z5, z6, z7, z8, z9 = z

    res['dy_dz8'] = 1  
    res['dy_dz9'] = 1  
    res['dz9_dz3'] = z6  
    res['dz9_dz6'] = z3 
    res['dz8_dz7'] = c1
    res['dz8_dc1'] = z7  
    res['dz7_dz1'] = z4
    res['dz7_dz4'] = z1
    res['dz6_dz5'] = c2
    res['dz6_dc2'] = z5
    res['dz5_dz2'] = 1 - np.tanh(z2)**2 
    sig_z2 = sigmoid(z2)
    res['dz4_dz2'] = sig_z2 * (1 - sig_z2)
    sig_z1 = sigmoid(z1)
    res['dz3_dz1'] = sig_z1 * (1 - sig_z1)
    res['dz2_dx2'] = 1
    res['dz2_db2'] = 1
    res['dz1_dx1'] = 1 
    res['dz1_db1'] = 1 

    return res
````

````python
answer['graph_derivatives'] = [graph_derivatives(**input) for input in inputs]
````

# Task 3

Using the values of the derivatives, calculated above:
- extract a dictionary of values that are needed to calculate $\partial_{c_1} y$
- calculate that derivative

````python
def dy_dc1(x, w):
    selected_derivs = {}
    dy_dc1 = 0

    ## YOUR CODE HERE ##
    graph = graph_derivatives(x, w)
    selected_derivs = {'dy_dz8': graph['dy_dz8'], 'dz8_dc1': graph['dz8_dc1']}
    dy_dc1 = graph['dy_dz8'] * graph['dz8_dc1']

    return selected_derivs, dy_dc1
````

````python
answer['dy_dc1'] = [dy_dc1(**input) for input in inputs]
````

# Task 4
Using the values of the derivatives, calculated above:
- extract a dictionary of values that are needed to calculate $\partial_{c_2} y$
- calculate that derivative

````python
def dy_dc2(x, w):
    selected_derivs = {}
    dy_dc2 = 0

    ## YOUR CODE HERE ##
    graph = graph_derivatives(x, w)
    
    selected_derivs = {'dy_dz9': graph['dy_dz9'], 'dz9_dz6': graph['dz9_dz6'], 'dz6_dc2' : graph['dz6_dc2']} # , 'dz9_dz3': graph['dz9_dz3']
    dy_dc2 = graph['dy_dz9'] * graph['dz9_dz6'] * graph['dz6_dc2'] # * graph['dz9_dz3']

    return selected_derivs, dy_dc2
````

# Task 5
Using the values of the derivatives, calculated above:
- extract a dictionary of values that are needed to calculate $\partial_{b_1} y$
- calculate that derivative

````python
def dy_db1(x, w):
    selected_derivs = {}
    dy_db1 = 0

    ## YOUR CODE HERE ##
    graph = graph_derivatives(x, w)
    
    selected_derivs = {
        # first path
        'dy_dz8': graph['dy_dz8'], 'dz8_dz7': graph['dz8_dz7'], 'dz7_dz1': graph['dz7_dz1'], 'dz1_db1': graph['dz1_db1'],
        # second path
        'dy_dz9': graph['dy_dz9'], 'dz9_dz3': graph['dz9_dz3'], 'dz3_dz1': graph['dz3_dz1']
    }
    path1 = graph['dy_dz8'] * graph['dz8_dz7'] * graph['dz7_dz1'] * graph['dz1_db1']
    path2 = graph['dy_dz9'] * graph['dz9_dz3'] * graph['dz3_dz1'] * graph['dz1_db1']

    dy_db1 = path1 + path2

    return selected_derivs, dy_db1
````

````python
answer['dy_db1'] = [dy_db1(**input) for input in inputs]
print(answer['dy_db1'])
````

# Task 6
Using the values of the derivatives, calculated above:
- extract a dictionary of values that are needed to calculate $\partial_{b_2} y$
- calculate that derivative

````python
def dy_db2(x, w):
    selected_derivs = {}
    dy_db2 = 0

    ## YOUR CODE HERE ##
    graph = graph_derivatives(x, w)

    selected_derivs = {
        # first path
        'dy_dz8': graph['dy_dz8'], 
        'dz8_dz7': graph['dz8_dz7'], 
        'dz7_dz4': graph['dz7_dz4'], 
        'dz4_dz2': graph['dz4_dz2'], 
        'dz2_db2': graph['dz2_db2'],
        # second path
        'dy_dz9': graph['dy_dz9'], 
        'dz9_dz6': graph['dz9_dz6'], 
        'dz6_dz5': graph['dz6_dz5'], 
        'dz5_dz2': graph['dz5_dz2']
    }
    
    path1 = graph['dy_dz8'] * graph['dz8_dz7'] * graph['dz7_dz4'] * graph['dz4_dz2'] * graph['dz2_db2']
    path2 = graph['dy_dz9'] * graph['dz9_dz6'] * graph['dz6_dz5'] * graph['dz5_dz2'] * graph['dz2_db2']

    dy_db2 = path1 + path2

    return selected_derivs, dy_db2
````

````python
answer['dy_db2'] = [dy_db2(**input) for input in inputs]
````
